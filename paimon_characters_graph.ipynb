```python
# -*- coding: utf-8 -*-
import matplotlib.pyplot as plt
import json

with open('paimon-moe-local-data.json','r') as f:
    json_file = json.load(f)
```

```python
characters_list = json_file.get('characters')
wish_list = []
for character in characters_list:
    print(characters_list[character])
    wish_list.append(characters_list[character]['wish'])
    print(wish_list)
plt.figure(figsize=(8, 16),dpi=100) #画像サイズ調整
plt.barh(range(len(wish_list)),wish_list) #横グラスのインスタンス
plt.grid(axis='x') #背景グリッド設定
plt.yticks(range(len(wish_list)),characters_list.keys()) #y軸設定
plt.show()

```

```python
import numpy as np
import matplotlib

matplotlib.rcParams['font.sans-serif']=['MS Gothic']
matplotlib.rcParams['axes.unicode_minus']=False
data = np.random.randn(1000)
plt.hist(data,bins=30)
plt.grid(axis='y') #背景グリッド設定
plt.title("ヒストグラフ")
plt.show()
```

```python
wish_count_event = json_file.get("wish-counter-character-event")["total"]
wish_count_standard = json_file.get("wish-counter-standard")["total"]
wish_count_weapon = json_file.get("wish-counter-weapon-event")["total"]
wish_count = wish_count_event + wish_count_standard + wish_count_weapon
data = [
    wish_count_weapon/wish_count*100,
    wish_count_standard/wish_count*100,
    wish_count_event/wish_count*100
]
plt.pie(data,labels=["武器祈願","常駐祈願","イベント祈願"],autopct="%1.1f%%")
plt.show()
```
